### Show Images Function

```python
# Function to display images side by side
def show_images(img1, title1, img2, title2, img3=None, title3=None):
    if img3 is not None and img3.shape[2] == 3:
        figsize = (14,8)
    elif img3 is not None and img3.shape[2] == 1:
        figsize = (12,12)
    elif img1.shape[1] < 100:
        figsize = (8,8)
    else:
        figsize = (15,15)
    
    if img3 is not None:
        fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize = figsize)
    else:
        fig, (ax1, ax2) = plt.subplots(1, 2, figsize = figsize)
        
    if (img1.shape[2] == 1):
        ax1.imshow(img1.squeeze(), cmap='gray')
        ax1.axis('off')
    else:
        ax1.imshow(img1)
        ax1.axis('off')
    ax1.set_title(title1, fontsize=20)

    if (img2.shape[2] == 1):
        ax2.imshow(img2.squeeze(), cmap='gray')
        ax2.axis('off')
    else:
        ax2.imshow(img2)
        ax2.axis('off')
    ax2.set_title(title2, fontsize=20)
        
    if img3 is not None:
        if (img3.shape[2] == 1):
            ax3.imshow(img3.squeeze(), cmap='gray')
            ax3.axis('off')
        else:
            ax3.imshow(img3)
            ax3.axis('off')
        ax3.set_title(title3, fontsize=20)

    if img3 is not None and img3.shape[2] == 3:
        fig.tight_layout()
    elif img1.shape[1] < 100:
        plt.tight_layout(pad=0.4, w_pad=10, h_pad=1.0)
```

### c) Creating Dataset to Use Multiple Camera Images

```python
sim_data_dir = 'C:/Users/gzx669/Documents/Udacity/Projects/Term1/Project3/data/'

# Extract all data from the simulation log file
rows = []
with open(sim_data_dir + 'driving_log.csv') as csvfile:
    log_csv = csv.reader(csvfile)
    next(log_csv, None)  # skip header row
    for row in log_csv:
        rows.append(row)
```

**Steering Angle Measurement Correction for Using Multiple Camera Images**
* In the `driving_log.csv`, the steering angle values are recorded along with center, left, and right camera images.
* In the left camera images, the car seems little off to the right of the road, and vice versa for the right camera images, while `steering` value may still 0.0.
* Thus, we can augment these images by **applying steering angle correction** and use the left and right camera images as if they were coming from the center camera.

```python
correction = 0.25

image_paths = []
steer_angles = []
samples = []

for row in rows:
    # Center camera images
    center = sim_data_dir + 'IMG/' + row[0].split('/')[-1]
    image_paths.append(center)
    steer_angles.append(float(row[3]))

    # Left camera images
    left = sim_data_dir + 'IMG/' + row[1].split('/')[-1]
    image_paths.append(left)
    steer_angles.append(float(row[3]) + correction)

    # Right camera images
    right = sim_data_dir + 'IMG/' + row[2].split('/')[-1]
    image_paths.append(right)
    steer_angles.append(float(row[3]) - correction)
    
print ('Length of image paths:    ', len(image_paths))
print ('Length of steering angles:', len(steer_angles))
    
samples = list(zip(image_paths, steer_angles))
```

## Step 1: Data Augmentation
**Flip Image** - The first training track is more left turn biased. Hence, in order to increase the generalization of the model, we flip images and the respective steering angle measurements.

```python
def flip_image(image, angle):
    return cv2.flip(image,1), angle*-1.0
```

```python
sample_flip, angle = flip_image(plt.imread(samples[i1][0]), samples[i1][1])
print('Original Steering Angle Value:',samples[i1][1])
print('Flipped Steering Angle Value:',angle)
show_images(plt.imread(samples[i1][0]), 'Original Image', sample_flip, 'Horizontally Flipped Image')
```

## Step 2: Image Preprocessing
**a) Adjust Brightness**
* We adjust brightness levels to random number of test images to avoid overfitting and increase generalization of the model.

**b) Apply Gaussian Blur**
* We apply Gaussian blurring to random number of test images to avoid overfitting and increase generalization of the model.

**c) Crop**
* The cameras in the simulator capture 160 x 320 pixel images. Not all of these pixels contain useful information, however. The top portion of the image captures trees and hills and sky, and the bottom portion of the image captures the hood of the car. Model might train faster if these pixels are cropped from each image to focus on only the portion of the image that is useful for predicting a steering angle. Thus, cropping 60 pixels from the top and 25 pixels from the bottom, and the resulting image resolution is 75 x 320 pixels.

**d) Resie**
* Resizing the images to 66 x 200 pixel resolution as per the Nvidia model.

**e) Change Color Space**
* Changing image color space from RGB to YUV as per the Nvidia model.

### a) Adjust Image Brightness

```python
def adj_brightness(image):
    brightness_coeff = 0.4
    # Change to HSV to change the brightness
    hsv = cv2.cvtColor(image, cv2.COLOR_RGB2HSV)
    hsv[:,:,2] = hsv[:,:,2] * brightness_coeff
    # Convert back to RGB
    return cv2.cvtColor(hsv, cv2.COLOR_HSV2RGB)
```

```python
sample_bright = adj_brightness(plt.imread(samples[i1][0]))
show_images(plt.imread(samples[i1][0]), 'Original Image', sample_bright, 'Brightness Adjusted Image')
```

### b) Gaussian Blur

```python
def gaussian_blur(image):
    return cv2.GaussianBlur(image, (3,3), 0)
```

```python
sample_blur = gaussian_blur(plt.imread(samples[i1][0]))
show_images(plt.imread(samples[i1][0]), 'Original Image', sample_blur, 'Gaussian Blurred Image')
```

### c) Crop Image

```python
def crop_img(image):
    top=60
    bottom=135
    return image[top:bottom]
```

```python
sample_crop = crop_img(plt.imread(samples[i1][0]))
show_images(plt.imread(samples[i1][0]), 'Original Image', sample_crop, 'Cropped Image')
```

### d) Resize Image

```python
def resize_img(image):
    return cv2.resize(image,(200,66), interpolation = cv2.INTER_AREA)
```

```python
sample_resize = resize_img(sample_crop)
show_images(plt.imread(samples[i1][0]), 'Original Image', sample_resize, 'Cropped Resized Image')
```

### e) Change Image Color Space from RGB to YUV

```python
def colorspace(image):
    return cv2.cvtColor(image, cv2.COLOR_RGB2YUV)
```

```python
sample_cspace = colorspace(sample_resize)
show_images(plt.imread(samples[i1][0]), 'Original RGB Image', sample_cspace, 'YUV Image')
```

### e) Function to Preprocess Image
* Apply brightness adjustment and Gaussian blur to random number of images
* Then crop and resize the images

```python
def preprocess(image, vis=False):
    
    # Probability for brighness adjustment
    prob_bright = np.random.randint(2)
    if prob_bright:
        bright = adj_brightness(image)
        if vis == True:
            print('Image Brightness Adjusted')
    else:
        bright = image
    
    # Probability for Gaussian blur
    prob_blur = np.random.randint(2)
    if prob_blur:
        blur = gaussian_blur(bright)
        if vis == True:
            print('Gaussian Blur Applied')
    else:
        blur = bright
    
    crop = crop_img(blur)
    
    resize = resize_img(crop)
    
    new_img = colorspace(resize)
    
    return new_img
```

**Visualize Preprocess Function on a Sample Image**

```python
sample_preprocess = preprocess(plt.imread(samples[i1][0]), True)
show_images(plt.imread(samples[i1][0]), 'Original Image', sample_preprocess, 'Preprocessed Image')
```

## Step 3: Generate Batches of Training and Validation Datasets
* Instead of building a large augmented and preprocessed dataset to feed to the model for training, we use `generator()` to apply image preprocessing techniques only to a small batch of images at a time while training the NN model.

```python
def generator(samples, batch_size=32):
    num_samples = len(samples)
    while 1: # Loop forever so the generator never terminates
        shuffle(samples)

        for offset in range(0, num_samples, batch_size):
            batch_samples = samples[offset:offset + batch_size]

            images = []
            angles = []
            
            for image_path, steer_angle in batch_samples:
                image = plt.imread(image_path)
                
                images.append(preprocess(image))
                angles.append(float(steer_angle))
                
                flipped_image, flipped_angle = flip_image(image, float(steer_angle))
                
                images.append(preprocess(flipped_image))
                angles.append(flipped_angle)
                
            X_train = np.array(images)
            y_train = np.array(angles)
            yield shuffle(X_train, y_train)
```

**Split Training and Validation Datasets**

```python
train_samples, valid_samples = train_test_split(samples, test_size=0.2)

train_generator = generator(train_samples, batch_size=int(32))
valid_generator = generator(valid_samples, batch_size=int(32))
```

## Step 4: Define the NN Model
* The model is based on NVIDIA's "End to End Learning for Self-Driving Cars" paper
* Source:  https://images.nvidia.com/content/tegra/automotive/images/2016/solutions/pdf/end-to-end-dl-using-px.pdf

```python
model1 = Sequential([
    # Normalize
    Lambda(lambda x: (x/ 127.5 - 1.),input_shape=(66, 200, 3)),
    
    # Three 5x5 convolution layers with maxpooling and dropouts
    Convolution2D(24, 5, 5, border_mode='same', activation='relu'),
    MaxPooling2D(),
    Dropout(0.2),
    Convolution2D(36, 5, 5, border_mode='same', activation='relu'),
    MaxPooling2D(),
    Dropout(0.2),
    Convolution2D(48, 5, 5, border_mode='same', activation='relu'),
    MaxPooling2D(),
    Dropout(0.2),
    
    # Two 3x3 convolution layers with maxpooling and dropouts
    Convolution2D(64, 3, 3, border_mode='same', activation='relu'),
    MaxPooling2D(),
    Dropout(0.2),
    Convolution2D(64, 3, 3, border_mode='same', activation='relu'),
    MaxPooling2D(),
    Dropout(0.2),
    
    # Flatten layer
    Flatten(),
    
    # Three fully connected layers (depth 100, 50, 10)
    Dense(100, activation='relu'),
    Dropout(0.2),
    Dense(50, activation='relu'),
    Dropout(0.2),
    Dense(10, activation='relu'),
    Dropout(0.2),
    
    # Fully connected output layer
    Dense(1),
    ])

model1.compile(loss='mse', optimizer=Adam(0.0001))
model1.summary()
```

```python
model2 = Sequential([
    # Normalize
    Lambda(lambda x: x / 127.5 - 1.0, input_shape=(66, 200, 3)),
    
    # Three 5x5 convolution layers (output depth 24, 36, and 48), each with 2x2 stride
    Convolution2D(24, 5, 5, border_mode='valid', subsample=(2, 2), W_regularizer=l2(0.001)),
    Activation('elu'),
    Convolution2D(36, 5, 5, border_mode='valid', subsample=(2, 2), W_regularizer=l2(0.001)),
    Activation('elu'),
    Convolution2D(48, 5, 5, border_mode='valid', subsample=(2, 2), W_regularizer=l2(0.001)),
    Activation('elu'),

    # Two 3x3 convolution layers (output depth 64, and 64)
    Convolution2D(64, 3, 3, border_mode='valid', W_regularizer=l2(0.001)),
    Activation('elu'),
    Convolution2D(64, 3, 3, border_mode='valid', W_regularizer=l2(0.001)),
    Activation('elu'),

    # Flatten layer
    Flatten(),

    # Three fully connected layers (depth 100, 50, 10)
    Dense(100, W_regularizer=l2(0.001)),
    Activation('elu'),
    Dense(50, W_regularizer=l2(0.001)),
    Activation('elu'),
    Dense(10, W_regularizer=l2(0.001)),
    Activation('elu'),

    # Fully connected output layer
    Dense(1),
    ])

model2.compile(loss='mse', optimizer=Adam(lr=0.0001))
model2.summary()
```

## Step 5: Train the NN Model

```python
history_object = model1.fit_generator(train_generator, samples_per_epoch = len(train_samples),
                                      nb_epoch=5, verbose = 1,
                                      validation_data=valid_generator,
                                      nb_val_samples=len(valid_samples))

model1.save('model1.h5')
print("\nModel Saved!")
```
